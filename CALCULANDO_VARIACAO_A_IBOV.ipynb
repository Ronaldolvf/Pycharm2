```python
import numpy as np
import pandas as pd
from pandas_datareader import data as wb
import datetime as dt
```

```python
dados = pd.read_csv('IBOVDiA3.csv')               # Importando os dados dp aquivo "IBOVDiA3.csv".
display(dados)
```

```python
tickers = dados["CD"]                            # Selecinando apenas os tickers das ações.
data_inicial = dt.datetime(2022,5,6)
data_final    = dt.datetime(2022,5,9)
lista = tickers #["VALE3.SA", "ASAI3.SA","ABEV3.SA","RRRP3.SA" ]
df = wb.DataReader (lista, "yahoo", data_inicial, data_final) 
df
```

```python
df['Close'] = df['Close'].pct_change()           # Calculando a variação diária da ação.

test = df[['Close']].round(4)*100    # Selecionando a coluna "Close" e multiplicando por 100 as cotações da linha 01.
                                     # Colocando 2 colchetes, cria-se um data freme. Essa função ocorre graças ao pandas. 
test
```

```python
ibov = test.iloc[1].round(3)              # Buscando os dados da linha 01.
sorted (ibov)                             # sorted é usado para colocar em ordem crescente
```

```python
test = pd.DataFrame(ibov)
test
```
